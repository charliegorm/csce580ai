```python
import time

def timed_test(missionaries_count,cannibals_count):
    test_agent = MCAgent()
    begin = time.time()
    solution = test_agent.solve(missionaries_count,cannibals_count)
    end = time.time()
    print(f"{missionaries_count}M, {cannibals_count}C -> moves={len(solution)} | solved={len(solution) > 0} | time={(end - begin):.6f}s")
    print(solution)
    
timed_test(1, 1)
timed_test(2, 2)
timed_test(3, 3)
timed_test(4, 3)
timed_test(5, 3)
timed_test(2, 3)
```
